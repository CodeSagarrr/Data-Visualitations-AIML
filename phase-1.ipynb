## Data Preprocessing (Visualization se pehle data ko tayyar karna)

* Data preprocessing matlab raw data ko clean, transform aur structure karna, taki visualization meaningful ho.

* Why important in AIML :

    - Outlier, missing values aur noise ko remove karna

    - Data ko normalize/standardize karna

    - Categorical data encode karna

    - Visualization ko accurate aur clear banana

* Common preprocessing steps :

    - Handling missing values

        - fillna(), dropna()

    - Outliers handle karna 

        - IQR method, z-score

    - Scaling

        - StandardScaler, MinMaxScaler

    - Encoding 

        - OneHotEncoding, Label Encoding

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("sample.csv")
df.head()
```

# Data Visualization Tools

## Matplotlib

* Low-level plotting library, jisme full customization milti hai.

* Usage:

    - Basic plots

    - Fine-grained control

    - AIML model results visualize karna (loss curves, accuracy curves)

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3] , [3,4,5])
plt.title("Simple Line Plot")
plt.xlabel("X Line")
plt.ylabel("Y Line")
plt.show()
```

# Seaborn

* Matplotlib ke upar built high-level library, quick statistical visualization.

* Usage:

    - Distribution plots

    - Correlation heatmap

    - Pairplots for ML dataset exploration

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(df["Index"])
plt.title("Id")
plt.show()
```

# Plotly

* Interactive, browser-based plotting library. 

* Usage:

    - Dashboard

    - Interactive ML visualizations

    - Time-series and 3D graphs

```python
import plotly.express as px

fig = px.histogram(x=[1, 2, 3], y=[1, 2, 3])
fig.show()
```
