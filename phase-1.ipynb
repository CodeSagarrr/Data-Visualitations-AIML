## Data Preprocessing (Visualization se pehle data ko tayyar karna)

* Data preprocessing matlab raw dataset ko clean, structured, consistent aur analysis-ready format me convert karna.
Real-world data hammesha messy hota hai. Isme missing values hoti hain, wrong formatting hoti hai, inconsistent units hote hain, aur kabhi-kabhi outliers bhi. Agar aap direct visualization kar doge, toh graphs misleading ho sakte hai.

* Why important in AIML :

    - ML model ko correct input mile

    - Visualization se sahi patterns dikhen

    - Scaling se features comparable bante hain

    - Encoding se categorical data numerically represent hota hai

* Important Tasks :

    - Handling missing values

        - fillna(), dropna()

    - Outliers handle karna 

        - IQR method, z-score

    - Scaling

        - StandardScaler, MinMaxScaler

    - Categorical Encoding 

        - OneHotEncoding, Label Encoding

    - Data Type Correction

    - Duplicate Handling

```python
import math
import random
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("sample.csv")

# Added Age column in Data frame
df["Age"] = [math.floor(random.randint(18 , 30)) for _ in range(len(df))]

# Missing values fill karna
# fillna(): missing values ko replace karta hai
df["Sex"] = df["Sex"].fillna("not included")

# Scaling numerical value
scaler = StandardScaler() 
# StandardScaler(): mean=0 aur variance=1 banata hai (z-score scaling)

df[["Index"]] = scaler.fit_transform(df[["Index"]])
# fit_transform(): pehle data pe parameters calculate karta hai, phir scale apply karta hai

# Final cleaned data dekhne ke liye
df.head()
```

# Data Visualization Tools

## Matplotlib

* Matplotlib Python ka sabse fundamental visualization library hai. Yeh ek low-level library hai jisme aap figure, axes, labels, ticks, markers sab manually customize kar sakte ho.
AIML me mostly training curves, feature distribution, confusion matrices Matplotlib se banaye jate hain.

* Usage:

    - Basic plots

    - Fine-grained control

    - AIML model results visualize karna (loss curves, accuracy curves)

```python
import matplotlib.pyplot as plt

# X and Y values define karna
X = [1, 2, 3]
Y = [4, 5, 6]

# plot(): x-y line plot draw karta hai
plt.plot(X , Y)

# title(): graph ka title set karta hai
plt.title("Simple Line Plot")

# xlabel(): X-axis ka label define karta hai
plt.xlabel("X axis")

# ylabel(): Y-axis ka label define karta hai
plt.ylabel("Y axis")

# grid(): graph me background grid lines add karta hai
plt.grid(True)

# show(): final graph render karta hai
plt.show()
```

# Seaborn

* Seaborn ek statistical visualization library hai jo matplotlib ke upar bani hai.
Yeh aesthetics improve karta hai aur statistical plots (distribution, correlation, pairplots) ko simple commands me convert kar deta hai. Machine learning me EDA (Exploratory Data Analysis) ke liye perfect tool hai.

* pairplot is a data visualization tool that shows pairwise relationships in a dataset by creating a matrix of scatterplots

* Correlation quantifies the linear relationship between two variables, ranging from -1 (perfect negative correlation) to +1 (perfect positive correlation), with 0 indicating no linear correlation.

* Usage:

    - Distribution plots

    - Correlation heatmap

    - Pairplots for ML dataset exploration

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(df["Index"])
plt.title("Id")
plt.show()
```

# Plotly

* Plotly ek interactive graph library hai jisme hover info, zooming, panning, selection sab built-in hota hai.
AIML dashboards ya model interpretation tools me kaafi useful hoti hai.

* Usage:

    - Dashboard

    - Interactive ML visualizations

    - Time-series and 3D graphs

```python
import plotly.express as px
import pandas as pd

df = pd.read_csv("sample.csv")

fig = px.histogram(df , x="Index" , y="Sex")
fig.show()
```

# Types of Plots

## Line Plot

* Line plot trend show karta hai, jaise time-series, model training loss, accuracy curves.


```python
loss = [0.8, 0.6, 0.4, 0.3]

# Agar x specify nahi karte toh x = index hota hai
plt.plot(loss)

plt.title("Training Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.grid(True);

plt.show()
```

## Bar Plot

* Category-wise comparison ke liye perfect hota hai.

```python
plt.bar(['A','B','C'] , [20,35,30])

# bar(): categorical data ko height ke form me show karta hai

plt.title("Category Count")
plt.show()
```

## Scatter Plot

* Relationship aur correlation identify karne ke liye.

```python
plt.scatter(df["Sex"] , df["Age"])
# scatter(): individual data points plot karta hai

plt.xlabel("Sex")
plt.ylabel("Index")
plt.show()
```

## Histogram

* Distribution samajhne me best.

```python
plt.hist(df["Age"] , bins=25)
# hist(): frequency distribution plot karta hai

plt.title("Salary Distribution")
plt.show()
```

# Visualization Best Practices (Expanded)

* Clarity
 
    - Graph simple rakho

    - Clear labels likho

    - Colors use karo, par zyada nahi

* Chart-Data Fit

    - Wrong chart â†’ wrong conclusion

    - Distribution â†’ histogram

    - Relationship â†’ scatter

    - Trend â†’ line

    - Categories â†’ bar

* Avoid Distortion

    - Axis ko unnecessarily zoom na karo

    - Viewers ko confuse na karo

* Annotate Important Points

    - Peak, drop, outliers highlight karo
